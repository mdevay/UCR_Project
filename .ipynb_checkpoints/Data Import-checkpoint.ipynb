## Initial Problem Statement/EDA

### Proposal:

Analysis of FBI Uniform Crime Reporting Data to determine trends in crime with regards to changes in Drug Policy

### Goal:

To determine if changes in Drug Policies have measurable impacts on crime rates and possibly other metrics. 

### Proposed Methods:

Datasets are released annually. The most recent will be read into python via pandas, and general characteristics will be assessed. Several cities with similar demographics will be selected from the data. A data dashboard will be created to navigate the raw data. Additional yearly data for the chosen cities will then be concatenated to the pandas dataset. Clustering or other categorization may be utilized. A model may be built to determine the impact of lessening or increasing restrictions on drug criminalization.

### Risks and Assumptions:

The UCR data is generated from voluntary reporting. Specific polices on which data to report vary somewhat between jurisdictions. Bias and/or skew may be present in the data due to this. Each jurisdiction may not report every crime to the database. Observation bias may be present.

### Success Criteria

1. Successful creation of a data dashboard/navigator
2. Establishment of trends within the data
3. Inference of impact to changes to drug policies
4. Modeling criteria and metrics will be determined prior to implementation

```python
# so the file is disjointed enough that using an apply method is not going to give us what we want in a single function. 
# we'll just do it with for loops instead.

def clean_up_table(df):
    '''Returns a new dataframe. Sets the column names, removes the first rows that are NA, and resets the index'''
    
#     new_df = pd.DataFrame()
#     new_df = df.drop(index=[0,1,2,3]).copy()
#     new_df.columns = df.loc[3,:]

    flag = 1
    counter = 0
    while flag:
        try:
            if df.iloc[counter, 0].strip() == 'State':
                flag = 0
                col_index = counter
            else:
                counter += 1
        except:
            counter += 1
    
    index = counter + 1
    new_df = df.loc[index:,:].copy()
    new_df.columns = df.loc[counter, :]
    new_df.index = range(len(new_df.index))
    
    return new_df
```

```python
# let's do something about those column names! We want them to be consistent from import to import
# but we also want them to be distinct. We'll have to rename all of them at the end anyways.
# lets just cut them to the first 8 characters or so and we'll rename them later.

def clean_column_names(df):
    '''removes columns without names, strips extra whitespace, and separates words with _ instead of a space for all columns names'''
    new_df = df.copy()
    
    try:
        new_df.columns = new_df.columns.fillna('remove_me')
        new_df = new_df.drop(columns='remove_me')
    except:
        pass
    
    cols = []
    for col in new_df.columns:
        
        col = "".join(re.findall("[A-Z, a-z]", col)).replace(" ", "")[:10].lower()
#         col = col.replace("-", "").lower()
#         col = "_".join(col.split())
        cols.append(col)
        
    new_df.columns = cols
    return new_df

ucr_clean = clean_column_names(ucr_clean)
ucr_clean.columns
```

```python
# now lets write a function that will put our data in a more useable format

def transform_table(df, year):
    '''consolidates relevant data into a streamlined dataframe'''
    
    new_df = pd.DataFrame(columns = df.columns)
#     new_df['year'] = year

    rows = len(df.index)
    indexer = 0
    new_df['population'] = np.nan
    check = df.columns[-1]
    df.columns = df.columns.fillna('check')
    
    # check to see the the excel format is of the newer or older type
    if "est" in check:
        new_df = transform_new_format(df, rows)
        new_df['year'] = year
        return new_df.astype(str)

    else:    
        for i in range(rows):
            row = df.loc[i, :]

            # if the row contains a state name, then get that name
            if get_state(row):
                state = get_state(row)

            # if the row has a population, then get that number
            elif get_pop(row):
                population = get_pop(row)

            # if the row contains the total data for the state, create a new row in the new df with the data for that state
            elif "Total" in str(row['state']):
                new_df.loc[indexer, :] = df.loc[i,:]
                new_df.loc[indexer, "state"] = state
                new_df.loc[indexer, "population"] = population
                # advance the index of the new dataframe by 1 for the next iteration of the loop
                indexer += 1

    new_df['year'] = year
              
    return new_df.astype(str)

def get_state(row):
    '''Takes a dataframe row and returns the name of a state if it is present'''
    
    try:
        if 'agen' in row['state']:
            state = "".join(re.findall("[A-Z]", row['state']))
            return state

    except:
        pass
   
def get_pop(row):
    '''Takes a dataframe row and returns the population of a state if it is present'''
    
    try:
        if 'population' in row['state']:
            population = "".join(re.findall("\d+", row['state']))
            return int(population)
    except:
        pass

def transform_new_format(df, rows):
    """This function is called if the excel spreadsheet uses the newer format"""
    
    new_df = pd.DataFrame(columns = df.columns)
    indexer = 0
    new_df['population'] = np.nan
    for i in range(rows):
        if df.iloc[i, 1] == np.nan:
            return new_df
        elif isinstance(df.iloc[i, 0], str):
            population = df.iloc[i, -1]
            state = "".join(re.findall("[A-Z]", str(df.iloc[i, 0])))
            
        else:
            new_df.loc[indexer, :] = df.loc[i,:]
            new_df.loc[indexer, 'population'] = population
            new_df.loc[indexer, 'state'] = state
            indexer += 1
    
    return new_df.astype(str)
```

```python
# now we just need to put it all together!

# lets make a function to import and clean the data for a single year:

def import_and_clean_UCR(year):
    path = "./data/ucr/UCR" + str(year) + ".xls"
    df = pd.read_excel(path)
    df = clean_up_table(df)
    df = clean_column_names(df)
    df = transform_table(df, year)
    return df
```

```python
# now lets make a function to import, clean, and concatenate multiple years

def create_ucr_df(years):
    '''input a list of years, and get a single cleaned up dataframe for those years'''
    old_df = import_and_clean_UCR(years[0])
    for i in range(1, len(years)):
        print(years[i])
        new_df = import_and_clean_UCR(years[i])     
        old_df = pd.merge(old_df, new_df, how='outer')
        print('success!!')
    old_df.index = range(len(old_df.index))
    return old_df
        
```

```python
# success!! a fully cleaned dataframe!
# now we need to append the drug policy data. 

# lets start by creating some new columns

ucr_final['decrim'] = 0
ucr_final['medical'] = 0
ucr_final['recreational'] = 0
```

```python
# basically, we've just created three one-hot encoding columns for decriminalization, 
# medical legalization, and recreational legalization
# now we need to add those values. unfortunately, I haven't been able to find
# a table with this information as such. So lets write a quick function that will 
# replace the values when we input the years:

def insert_policy_flags(juris, yr_decrim=0, yr_med=0, yr_rec=0, df=ucr_final):
    state = df['state'] == juris
    max_year = df[state]['year'].max()

    if yr_rec:
        if not yr_decrim:
            yr_decrim = yr_rec
    def update_policy(year, policy):
        if year:
            for i in range(year, max_year + 1):
                year = df['year'] == i
                filt = year & state
                try:
                    df.loc[filt, policy] = 1
                except:
                    pass
    
    update_policy(yr_decrim, "decrim")
    update_policy(yr_med, "medical")
    update_policy(yr_rec, "recreational")
    
        

insert_policy_flags('ALASKA', yr_decrim=2003, yr_med=1998, yr_rec=2014)
```
